```python
import numpy as np

zero = np.array([[1],
                [0]])

X = np.array([[0,1],
             [1,0]])

print(X @ zero)
```

```python

zero = np.array([[1],
                 [0]])

one = np.array([[0],
                [1]])

plus = (1/np.sqrt(2)) * np.array([[1],
                                  [1]])

print(zero)
print(one)
print(plus)
```

```python
wrong = np.array([[1],
                  [1]])

print(np.linalg.norm(wrong)) # L2 norm 반환 및 출력
```
